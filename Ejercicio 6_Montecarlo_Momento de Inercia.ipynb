```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def correlacion_k_30(N=10000, k=30):
    valores = np.random.rand(N)
    lista = np.array([])
    for i in range(1,k+1):
        suma=0
        for j in range(N-k):
            suma += valores[j]*valores[j+i]
        lista = np.append(lista,suma)
    lista_ = lista/N
    return lista_
```

```python
plt.plot(correlacion_k_30())
plt.xlim(-2,31)
plt.ylim(0.2,0.3)
plt.grid()
plt.xlabel("k-esimo vecino")
plt.ylabel("CK")
plt.show()
```

```python
#Se puede apreciar que el valor está fluctuando alrededor del valor teórico (1/4 = 0.25)
```
