```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = lambda x: np.exp(x)*(x**3/(np.exp(x)-1))
rta = np.pi**4/15
```

```python
def integral_laguerre(f, n):
    puntos,pesos=np.polynomial.laguerre.laggauss(n)
    return np.sum(pesos*f(puntos))

# Evaluada en n=3
integral_n3 = integral_laguerre(f,3)
print(integral_n3)
```

```python
#error
x=np.arange(2,11,1)
y=np.zeros(0)
for n in x:
    int_= integral_laguerre(f,n)
    error= int_/rta
    y=np.append(y,error)
```

```python
plt.grid()
plt.scatter(x,y,label="Laguerre quadrature acurrency")
plt.legend(loc=0)
plt.show()
```
