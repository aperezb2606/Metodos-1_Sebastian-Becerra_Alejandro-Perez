```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p1 = np.linspace(0.1, 0.9, 100)
p2 = np.linspace(0.1, 0.5, 100)


P1, P2 = np.meshgrid(p1, p2)


probability = P1 * P2 * (1 - P1) * (1 - P2) + P1 * (1 - P2) * (1 - P1) * P2
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(P1, P2, probability, cmap='viridis')
ax.set_xlabel('p1')
ax.set_ylabel('p2')
ax.set_zlabel('Probabilidad')
ax.set_title('Superficie de Probabilidad del Evento A')
plt.show()
```

```python
min_probability = np.min(probability)
max_probability = np.max(probability)
min_index = np.unravel_index(np.argmin(probability), probability.shape)
max_index = np.unravel_index(np.argmax(probability), probability.shape)
min_p1 = p1[min_index[1]]
min_p2 = p2[min_index[0]]
max_p1 = p1[max_index[1]]
max_p2 = p2[max_index[0]]
```
