```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lista=[]
for i in range(81): 
    x=(np.math.factorial(365))/((365**i)*(np.math.factorial(365-i)))
    lista.append(x)

x= np.arange(0,81,1)
plt.figure()
plt.xlabel('personas')
plt.ylabel("Pn")
plt.plot(x,lista)
plt.show()
```
