```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def calculate_probability(n):
    numerator = math.factorial(365)
    denominator = (365 ** n) * math.factorial(365 - n)
    probability =1- (numerator / denominator)
    return probability
print(calculate_probability(23))
```

```python
Pn=np.zeros(80)
for i in range(1,81):
    Pn[i-1]= calculate_probability(i)

x=np.linspace(1,80,80)   
plt.plot(x,Pn)
plt.title('Probabilidad de coincidir en la fecha de cumpleaños según la cantidad de personas')
```
