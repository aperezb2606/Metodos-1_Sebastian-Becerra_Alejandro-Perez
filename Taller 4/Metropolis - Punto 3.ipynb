```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def distribucion_gaussiana(x, valor_dist_gauss, std):
    return (1 / (std * (2 * np.pi) ** 0.5)) * np.exp(-(x - valor_dist_gauss) ** 2 / (2 * std ** 2))

N = 1000
i = 0
valor_dist_gauss = 2.0
desv_estandar = 0.5
delta = 1
aceptacion = 0
muestras = np.zeros(N)
muestras[0] = 1.01

```

```python
while i < N - 1:
    y = muestras[i] + np.random.uniform(-delta, delta)
    alfa = min(1, distribucion_gaussiana(y, valor_dist_gauss, desv_estandar) / distribucion_gaussiana(muestras[i],  valor_dist_gauss, desv_estandar))
    r = np.random.rand()
    
    if r < alfa:
        muestras[i + 1] = y
        aceptacion += 1
        i += 1
    else:
        muestras[i + 1] = muestras[i]
        i += 1

porcentaje_aceptacion = (aceptacion / N) * 100
print("{0}{1}".format(porcentaje_aceptacion, "%"))
```
