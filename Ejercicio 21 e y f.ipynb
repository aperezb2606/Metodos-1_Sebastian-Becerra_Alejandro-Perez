```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
from tqdm import tqdm
```

```python
def legendre(n):

    if n == 0:
        return np.poly1d([1])
    elif n == 1:
        return np.poly1d([1, 0])
    else:
        p_n_1 = legendre(n-1)
        p_n_2 = legendre(n-2)
        return ((2*n-1)*np.poly1d([1, 0])*p_n_1 - (n-1)*p_n_2) / n


for n in range(21):
    print(f"n = {n}: {legendre(n)}")
```

```python
def legendre_C(n):

    if n == 0:
        return np.poly1d([1])
    elif n == 1:
        return np.poly1d([1, 0])
    else:
        p_n_minus_1 = legendre(n-1)
        p_n_minus_2 = legendre(n-2)
        return ((2*n-1)*np.poly1d([1, 0])*p_n_minus_1 - (n-1)*p_n_minus_2) / n


for n in range(20):
    p = legendre_C
    p_0 = p(0)
    print(f"c{n}(0) = {p_0}")
```
