```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
funcion = lambda x: (np.tan(x))**(1/2)
h = 0.01
valores_x = np.arange(0.1, 1.1, h)
```

```python
# Enunciado c:
def D_progresiva(f,x,h):
    return (1/(2*h))*(-3*f(x)+4*f(x+h)-f(x+2*h))

print("Derivadas progresivas")
print(D_progresiva(funcion,valores_x, h)) 

# Enunciado D:

def D_central(f,x,h):
    return (f(x+h)-f(x-h))/(2*h)

print("Derivadas centrales")
print(D_central(funcion,valores_x, h)) 
```

```python
#Enunciado e:
derivada_f = lambda x: (1/((np.cos(x))**2))*(1/(2*((np.tan(x))**(1/2))))

print(derivada_f(valores_x))
```

```python
plt.plot(valores_x,D_progresiva(funcion,valores_x,h), color = "r", label='Derivada progresiva')
plt.plot(valores_x,D_central(funcion,valores_x,h), color ="g", label='Derivada central')
plt.plot(valores_x,derivada_f(valores_x), color ="b", label='Derivada exacta')
plt.legend(loc=0)
plt.show()

# En principio, parece que, a comparación de la derivada exacta, la derivada progresiva y central tienen la misma precisión.
```

```python
#Enunciado g:

#Errores nodales:

error_progresiva = np.abs(derivada_f(valores_x) - D_progresiva(funcion,valores_x,h))
error_central = np.abs(derivada_f(valores_x) - D_central(funcion,valores_x,h))
```

```python
plt.plot(valores_x,error_progresiva, color = "r", label='Error derivada progresiva')
plt.plot(valores_x,error_central, color ="g", label='Error derivada central')
plt.legend(loc=0)
plt.show()
```

Se puede ver que ambas aproximaciones, tanto la central como la progresiva, son muy buenas ya que ambas tienen un error nodal muy pequeño con respecto a la derivada exacta, tal que no se aprecia en la primera gráfica. No obstante, al ver la gráfica de los errores podemos apreciar que las aproximaciones no poseen el mismo orden de precisión, siendo la derivada central más precisa que la derivada progresiva, ya que posee un error menor (ver eje y de la gráfica de errores).
