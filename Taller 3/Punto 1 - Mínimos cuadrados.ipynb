```python
import numpy as np
import matplotlib.pyplot as plt
```

ENUNCIADO A

```python
# Ecuaciones de las líneas en R^2:
#l1: 2x − y = 2
#l2: x + 2y = 1
#l3: x + y = 4

def minimos_cuadrados(A, b):
    sol = np.linalg.inv(A.T @ A) @ A.T @ b
    return sol

A = np.array([[2, -1], [1, 2], [1, 1]]) 
b = np.array([2, 1, 4]) 

sol = minimos_cuadrados(A,b)

print(sol)
```

```python
# Despejamos y de todas las ecuaciones de las lineas:
#l1: y = 2x-2
#l2: y = (1-x)/2
#l3: y = 4 -x

x = np.linspace(-5, 5, 100)
y1 = 2*x-2
y2 = (1-x)/2
y3 = 4-x

plt.plot(x, y1, label='2x - y = 2')
plt.plot(x, y2, label='x + 2y = 1')
plt.plot(x, y3, label='x + y = 4')
plt.scatter(sol[0], sol[1], color="r", label='Punto')
plt.legend(loc=0)
plt.grid()
plt.show()
```

Es posible interpretar el punto obtenido como la intersección de las tres líneas, que aunque no parece que se intersecan en la gráfica, si lo hacen. Ya con el método de mínimos cuadrados logramos encontrar uno de los puntos de intersección.

ENUNCIADO B

```python
h = 0.01
x = np.arange(-5, 5, h)
y = np.arange(-5, 5, h)
z = np.zeros((len(x), len(y)))

A = np.array([[2, -1], [1, 2], [1, 1]]) 
b = np.array([2, 1, 4]) 

def distancia(punto, A, b):
    return np.linalg.norm(np.dot(A, punto) - b)

i = 0
while i < len(x):
    j = 0
    while j < len(y):
        punto = np.array([x[i], y[j]])
        distancia_ = distancia(punto, A, b)
        z[i][j] = distancia_
        j += 1
    i += 1
```

```python
fig = plt.figure()
ejes = fig.add_subplot(111, projection='3d')
x_, y_ = np.meshgrid(x, y)
ejes.plot_surface(x_, y_, z, cmap='coolwarm')
ejes.set_xlabel('X')
ejes.set_ylabel('Y')
ejes.set_zlabel('Distancia')
ejes.view_init(elev=20, azim=35)
ejes.scatter(sol[0], sol[1], color="r", label='Punto')
plt.show()
```
