```python
import numpy as np
import sympy as sym
```

```python
def laguerre_(n):
    
    x = sym.Symbol('x',Real=True)
    y = sym.Symbol('y',Real=True)
    
    y = sym.exp(-x)*x**n
    
    polinomio = sym.exp(x)*sym.diff(y,x,n)/(np.math.factorial(n))
    polinomio = sym.expand(polinomio)
    return polinomio
```

```python
laguerre_(0) #Se puede ver que con n=0 el resultado es 1, que no es un polinomio ni tiene raices
```

```python
def laguerre20():
    Laguerre=[]
    for i in range(1,21):
        Laguerre.append(laguerre(i))
    return Laguerre

print(laguerre20())
```

```python
def raices_laguerre():
    raices = []
    n = 1
    while n <= 20:
        r = sym.poly(laguerre_(n)).all_roots()
        xi = [rt.evalf(10) for rt in r]
        raices.append(xi)
        n+=1
    return raices

print(raices_laguerre())
```

```python
pol_raiz = []
for i in range(20):
    pol = laguerre20()[i]
    raiz = raices_laguerre()[i]
    a = ["Polinomio de orden {0}: ({1}) --> Raices: {2}".format(i+1,pol,raiz)]
    pol_raiz.append(a)
print(pol_raiz)
```
